# Poisson model of neurons.
We start with a Poisson model to describe the spiking of neurons in ACC. This model is accompanied with a theory for rates, covairances and correlations. Therefore we need not run simulations of the network yet. Release of NE is modeled by modulation in attentional gain as in Ecker 2015.

```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
from sympy import symbols, Eq, solve
from sympy import Matrix
import scipy as sp
import seaborn as sns
```

```python
# Number of neurons.
N = 1000
np.random.seed(0)
# Rates of each neuron.
# Rates follow a log normal distribution.
f = np.random.lognormal(1,1,(N,1))#.reshape(N,1) # Hz or sp counts?
# Or rates are according to a stimulus preference.
# theta = 0
# phi = np.linspace(-np.pi,np.pi,N).reshape(N,1)
# epsilon = 0; kappa=2
# f = np.exp(kappa * np.cos(theta-phi)+epsilon)
mu = 0.1 # Mean gain.
sigma = 0.1 # Std of gain.


# Start the figure.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.9, rc={"lines.linewidth":3.3})
plt.axvline(x = np.mean(f))

# histogram of rates.
sns.distplot(f,kde = False)
# plt.plot(phi,f)
plt.xlabel(r'$f =$ rate')

sns.despine()

plt.show()
```

```python
np.mean(f)
```

```python
def correlation_from_covariance(covariance):
    v = np.sqrt(np.diag(covariance))
    outer_v = np.outer(v, v)
    correlation = covariance / outer_v
    correlation[covariance == 0] = 0
    return correlation
```

```python
Cov = mu * np.diag(f[:,0]) + sigma**2 * f @ np.transpose(f)
```

```python
# Plot of sample correlation matrix.
# Start the figure.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.9, rc={"lines.linewidth":3.3})

Corr = correlation_from_covariance(Cov)
plt.imshow(Corr)
plt.colorbar()

plt.xlabel(r'$\phi$')

plt.ylabel(r'$\phi$')

sns.despine()

plt.show()
```

```python
# Plot distribution of Fano factor.
# Start the figure.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.9, rc={"lines.linewidth":3.3})

# Fano factor.
Fano = 1 + sigma**2 / mu * f

sns.distplot(Fano)

plt.xlabel(r'Fano')
plt.ylabel('Count')

sns.despine()

plt.show()
```

```python
# Here we evaluate the mean correlations and mean Fano factor as a function of mu, sigma, and rates.


# Number of neurons.
N = 1000
np.random.seed(0)
# Rates of each neuron.
f = np.random.lognormal(1,1,(N,1))#.reshape(N,1) # Hz or sp counts?
# theta = 0
# phi = np.linspace(-np.pi,np.pi,N).reshape(N,1)
# epsilon = 0; kappa=2
# f = np.exp(kappa * np.cos(theta-phi)+epsilon)

Num_points_mu = 20
Num_points_sigma = 20
mu = np.linspace(0.1,0.25,Num_points_mu)
sigma = np.linspace(0.05,0.15,Num_points_sigma)

Avg_corr = np.zeros((Num_points_mu,Num_points_sigma))
Fano_mean = np.zeros((Num_points_mu,Num_points_sigma))

for i in range(Num_points_mu):
    for j in range(Num_points_sigma):
        Cov = mu[i] * np.diag(f[:,0]) + sigma[j]**2 * f @ np.transpose(f)
        Corr = correlation_from_covariance(Cov)
        a = Corr[np.triu(Corr, k = 1)!=0]
        # Mean correlations
        Avg_corr[i,j] = np.mean(a)
        # Fano factor.
        Fano_mean[i,j] = np.mean(1 + sigma[j]**2 / mu[i] * f)
        
        
f = np.random.lognormal(0.7,1,(N,1))#.reshape(N,1) # Hz or sp counts?
Avg_corr_lowf = np.zeros((Num_points_mu,Num_points_sigma))
Fano_mean_lowf = np.zeros((Num_points_mu,Num_points_sigma))

for i in range(Num_points_mu):
    for j in range(Num_points_sigma):
        Cov = mu[i] * np.diag(f[:,0]) + sigma[j]**2 * f @ np.transpose(f)
        Corr = correlation_from_covariance(Cov)
        a = Corr[np.triu(Corr, k = 1)!=0]
        # Mean correlations
        Avg_corr_lowf[i,j] = np.mean(a)
        # Fano factor.
        Fano_mean_lowf[i,j] = np.mean(1 + sigma[j]**2 / mu[i] * f)
```

```python
# Start the figure.
fig=plt.figure(figsize=(8,5))
ax = plt.subplot(111)

sns.set()
sns.set_style('whitegrid')
sns.set_style('white')
sns.set_style("ticks")
sns.set_context("talk", font_scale=1.9, rc={"lines.linewidth":3.3})

# # Plot previous results.
# plt.imshow(Avg_corr)
# plt.colorbar()

ax = sns.heatmap(Avg_corr,
                yticklabels= (0.1,0.25),
                xticklabels= (0.05,0.15),
                rasterized=True,
                vmin=0,vmax=0.4
                )
plt.xticks((0,Num_points_mu-1))
plt.yticks((0,Num_points_sigma-1))
plt.yticks(rotation=0)
plt.xticks(rotation=0)
ax.tick_params(axis='both', which='both', length=0)
# plt.axvline(x=7, color='white',linestyle='--',linewidth=1.5)
# plt.axvline(x=16, color='white',linestyle='--',linewidth=1.5)

cbar = ax.collections[0].colorbar
cbar.set_ticks([0,0.2,0.4, 0.6])
# cbar.set_ticklabels(['low', '20%', '75%', '100%'])
cax = plt.gcf().axes[-1]
cax.tick_params(size=0)




plt.ylabel(r'$\mu$')
plt.xlabel(r'$\sigma$')

plt.title('Mean rsc',fontsize=18)

# fig.savefig("./PreliminaryFigures/corrsHeatmap.pdf", bbox_inches="tight")

plt.show()
```

How can we have the Fano factor and correlations follow opposite trends? Is it possible with this simple model?

```python
# Make heatmaps of correlations and Fano as a fucntion of sigma and f.

# Number of neurons.
N = 1000
np.random.seed(0)
# Rates of each neuron.
f = np.random.lognormal(1,1,(N,1))#.reshape(N,1) # Hz or sp counts?
# theta = 0
# phi = np.linspace(-np.pi,np.pi,N).reshape(N,1)
# epsilon = 0; kappa=2
# f = np.exp(kappa * np.cos(theta-phi)+epsilon)

mu = 0.1

Num_points_f = 20
Num_points_sigma = 20
mean_f = np.linspace(0.75,1.25,Num_points_f)
sigma = np.linspace(0.05,0.15,Num_points_sigma)

Avg_corr = np.zeros((Num_points_f,Num_points_sigma))
Fano_mean = np.zeros((Num_points_f,Num_points_sigma))

for i in range(Num_points_f):
    f = np.random.lognormal(mean_f[i],1,(N,1))#.reshape(N,1) # Hz or sp counts?
    for j in range(Num_points_sigma):
        Cov = mu * np.diag(f[:,0]) + sigma[j]**2 * f @ np.transpose(f)
        Corr = correlation_from_covariance(Cov)
        a = Corr[np.triu(Corr, k = 1)!=0]
        # Mean correlations
        Avg_corr[i,j] = np.mean(a)
        # Fano factor.
        Fano_mean[i,j] = np.mean(1 + sigma[j]**2 / mu * f)
        
        
```
